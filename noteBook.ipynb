### Metodo númerico para encontrar el valor del ángulo de disparo:

```python
import numpy as np

# Definir las constantes
#-------------------------------------------------------
Potencia = 1500
R        = 3 
Vm       = 177
#-------------------------------------------------------
k = Vm**2 / (2 * np.pi * R)

res0 = Potencia/k 

angles = np.linspace(0, 2*np.pi, 10000000) #10 millones de muestras entre 0 y 2pi radianes
defAns = None

for i, ang in enumerate(angles):
    res = np.pi -ang + np.cos(ang)*np.sin(ang)
    
    if i==0 :   # 1era iteración
        err =  abs(res-res0)
        
        defAns = ang    #Angulo de Disparo
        continue
    
    currErr = abs(res-res0) #Obtener el error
    
    if currErr < err:   #Si el error de un angulo determinado es menor al error anterior, 
                        #hacer este la nueva mejor aproximación
        err = currErr
        defAns = ang
```

```python
print("Ángulo de disparo estimado: ")
print("-------------------------------")
print("Radianes:", defAns)
print("Grados", np.degrees(defAns),"°")
```

```python
print("Ángulo de disparo + pi en: ")
print("-------------------------------")
a1=defAns+np.pi
print("Radianes:", a1)
print("Grados", np.degrees(a1),"°")
```

```python
print("Resultado con el ángulo de disparo estimado: ")
np.pi -defAns -np.cos(defAns)*np.sin(defAns)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
frecuencia = 1  # Frecuencia en Hz
amplitud = 1    # Amplitud
periodo = 1 / frecuencia  # Período
t_muestreado = np.linspace(0, periodo, 9)   # Tiempo muestreado (6 intervalos)
señal_muestreada = amplitud * np.sin(2 * np.pi * frecuencia * t_muestreado)**2  # Señal muestreada

# Generar los valores para las barras en medio de los puntos
t_barras = np.zeros(2 * len(t_muestreado))  # Tiempos para las barras
valores_barras = np.zeros(2 * len(t_muestreado))  # Valores repetidos para las barras

t_barras[0::2] = t_muestreado
t_barras[1::2] = np.roll(t_muestreado, -1)  # Mover tiempos un paso adelante
valores_barras[0::2] = señal_muestreada
valores_barras[1::2] = señal_muestreada  # Repetir los valores de las muestras

# Graficar
plt.figure(figsize=(8, 5))

# Reconstrucción en escalón
plt.step(t_barras[:-1], valores_barras[:-1], where='post', color='orange', label="Reconstrucción (escalón)")

# Puntos muestreados
plt.scatter(t_muestreado, señal_muestreada, color='red', zorder=5, label="Puntos muestreados")

# Personalización del gráfico
plt.title("Reconstrucción de señal sinusoidal con barras entre puntos")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np

# Muestras de la señal (simulada o medida)
x = np.array([1, 0.87, 0.5, 0, -0.5, -0.87, -1, -0.87, -0.5])  # Ejemplo

# Aplicar DFT
N = len(x)
X = np.fft.fft(x)

# Calcular amplitudes reales
A = 2 * np.abs(X[:N//2]) / N  # Dividimos hasta la mitad de frecuencias

# Extraer amplitudes
A1 = A[1]  # Fundamental
Ak = A[2:]  # Armónicos

# Calcular THD
THD = np.sqrt(np.sum(Ak**2)) / A1 * 100
print(f"THD: {THD:.2f}%")
```
